#Step 3: Analyse Dataframe To get Our Result 

**Getting College name in the First Row**

```
p=df['Unnamed: 0']
```

**We are to select Colleges of Savitribai Phule Pune University and Civil Engineering Brance Only**

```
count=(p.str.contains('Savitribai Phule Pune University')&p.str.contains('Civil Engineering')).sum()
if count>0:
    print ("There are {m} Matches Found".format(m=count))
```

**Getting those indices from the First Feature and we need indices for next two Rows**

**As the we observed next two rows of the 2nd column contains GOPENS and GOPENH category and there Cut-off Rank(Cut-off marks)**

```
idx=p.index[(p.str.contains('Savitribai Phule Pune University')&p.str.contains('Civil Engineering'))== True].tolist()
idx_1=[x + 1 for x in idx] 
idx_2=[x + 2 for x in idx] 
```

```
df_col=df.loc[ idx , : ]
df_cat=df.loc[ idx_1, : ]
df_cut=df.loc[idx_2, : ]
```

```
frames=[df_col,df_cat,df_cut]
df_new=pd.concat(frames)
```

```
df_new=df_new.sort_index(axis=0)
```

```
df_new
```

**Dropping Those Columns Which are all Empty**

```
df_new=df_new.dropna(axis=1,thresh=1)
df_new=df_new.fillna("##")
```

```
df_new.reset_index(drop=True, inplace=True)
```

```
df_new.columns
```

```
df_new
```

**Rename Columns**

```
df_new=df_new.rename(columns={'Unnamed: 0':'a1', 'Unnamed: 1':'a2', 'Unnamed: 3':'a3', 'Unnamed: 5':'a4', 'Unnamed: 7':'a5',
       'Unnamed: 9':'a6', 'Unnamed: 10':'a7', 'Unnamed: 11':'a8', 'Unnamed: 12':'a9',
       'Unnamed: 13':'a10', 'Unnamed: 14':'a11', 'Unnamed: 15':'a12', 'Unnamed: 16':'a13',
       'Unnamed: 17':'a14', 'Unnamed: 18':'a15', 'Unnamed: 19':'a16', 'Unnamed: 20':'a17',
       'Unnamed: 21':'a18', 'Unnamed: 22':'a19', 'Unnamed: 23':'a20', 'Unnamed: 25':'a21',
       'Unnamed: 26':'a22', 'Unnamed: 27':'a23'})
```

```
df_t=df_new.T
df_t
```

For GOPENH and GOPENS we need only First two columns a1,a2

```
df_result=df_new[['a1','a2']]
```

```
df_result['a1'].unique()
```

```
df_result.shape
```

**We are to Find out Only the indices of GOPEN and GOPENS category , since it may be different**

```
index = df_result.index
condition1= df_result['a2'] == 'GOPENH'
condition2=df_result['a2'] == 'GOPENS'
```

```
indices1 = index[condition1]
indices2 = index[condition2]
indices_list1 =indices1.tolist()
indices_list2 =indices2.tolist()
```

```
indiceslist=indices_list1+indices_list2
```

```
indicesresult=[x+1 for x in indiceslist]
indi=[x-1 for x in indiceslist]
```

```
indicesresult.sort()
indi.sort()
```

```
len(indicesresult)
```

**We are to find out corresponding cut-offs**

```
df_cutoff=df_result.loc[indicesresult, : ]
```

```
df_res=df_cutoff['a2']
```

**DataFrame contains both Cutt-off ranks and Cut-off Marks**

```
df_res
```

```
import re
lst=[]
for i in df_res:
  result=re.search('(.*)\n',i)
  x=result.group(1)
  lst.append(x)
df_cutof = pd.DataFrame(lst,columns =['CutOff_Rank']) 
```

**Dataframe For Cut-off Ranks**

```
df_cutof
```

```
def find_between( s, first, last ):
    try:
        start = s.index( first ) + len( first )
        end = s.index( last, start )
        return s[start:end]
    except ValueError:
        return ""
lst=[]
for i in df_res:
  result=find_between(i, '\n(',')')
  lst.append(result)
df_cutof_marks = pd.DataFrame(lst,columns =['CutOff_Marks']) 
```

**Dataframe for Cut-off Marks**

```
df_cutof_marks
```

**We are to find out only college name fron first column**

```
df_colname=df_result.loc[indi, : ]
```

```
df_rescol=df_colname['a1']
```

```
df_rescol
```

```
lst=[]
for i in df_rescol:
  result=re.search('- (.*)\n',i)
  x=result.group(1)
  lst.append(x)
df_colnamee= pd.DataFrame(lst,columns =['College_Name']) 
```

Dataframe of only **College Name**

```
df_colnamee
```

```
df_re=pd.concat([df_colnamee,df_cutof,df_cutof_marks],axis=1)
```

#Step4: Dataframe contains College Name, Cut-off Ranks and Cut-off Marks Only of GOPENH/GOPENS of Civil Engineering Branch of Savitribai Phule Pune University 

```
df_re
```

```
df_re.CutOff_Rank=df_re.CutOff_Rank.astype(int)
```

```
df_re.CutOff_Rank.dtype
```

```
res = sorted(range(len(df_re.CutOff_Rank)), key = lambda sub: df_re.CutOff_Rank[sub])[:10]
print("Indices list of min 10 elements is : " + str(res))  
```

```
df_re_final=df_re.loc[res, : ]
```

```
df_re_final.reset_index(drop=True, inplace= True)
```
